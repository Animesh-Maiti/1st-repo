```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
A = np.random.rand(10,10)
plt.figure(figsize =(6,5))
plt.imshow(A,cmap = 'hot', interpolation = 'nearest')
plt.colorbar(label = "industry")
plt.title("heatmap of A")
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.show()
```
