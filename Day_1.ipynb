```python
import matplotlib.pyplot as plt

x=[1,2,3,4,5,6]
y=[1,4,9,16,25,36]

plt.plot(x,y)

plt.title("square number")
plt.grid(True)
plt.legend()
plt.show()
```

```python
plt.plot(x,y,'ro--',label='growth')
```

```python
labels =['A','B','C','D']
scores =[90,80,70,60]
colors =['red','green','blue','yellow']
plt.bar(labels,scores,color = colors , width = 0.8)
plt.ylim(0,100)
plt.title("exam marks")
plt.show()
```

```python
labels = ['rent', 'food', 'transport','savings']
values = [40,20,15,25]

plt.pie(values, labels = labels)
plt.title("monthly expenditure")
plt.show()
```
